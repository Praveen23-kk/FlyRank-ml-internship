## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Unique Discovery: Does high keyword competition lead to higher engagement rates?
# Use .copy() to avoid SettingWithCopyWarning
engaged_pages = df[df['engagement_rate'] > 0].copy()
comp_engagement_corr = engaged_pages['competition'].corr(engaged_pages['engagement_rate'])

print(f"Correlation between Competition and Engagement Rate: {comp_engagement_corr:.3f}")

# Visualizing the relationship via a binned average
# Explicitly handle duplicates and observed parameter for future-proofing
engaged_pages['comp_bin'] = pd.qcut(engaged_pages['competition'], q=5, duplicates='drop')
engagement_by_comp = engaged_pages.groupby('comp_bin', observed=False)['engagement_rate'].mean()

print("\nAverage Engagement Rate by Competition Quintile:")
print(engagement_by_comp.round(2))

engagement_by_comp.plot(kind='line', marker='o', title='Engagement Rate vs Competition Quintile')
plt.ylabel('Mean Engagement Rate (%)')
plt.xlabel('Competition Score (Low to High)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import shap
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Load the processed features
features_df = pd.read_csv("data/processed/refresh_feature_vector.csv")

# Based on the previous error output, the target column is 'is_declining_label'
target_col = 'is_declining_label'

# Separate features (X) and label (y)
y = features_df[target_col]
X = features_df.drop(columns=[target_col, 'content_id', 'client_id'], errors='ignore')

# Ensure only numeric columns are used for the SHAP analysis
X = X.select_dtypes(include=['number'])

# Train a model to explain
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X, y)

# Calculate SHAP values
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

# Handle SHAP output format variations (v0.45+ vs older)
if isinstance(shap_values, list):
    sv = shap_values[1] # For binary classifier, take positive class
else:
    # If it's a single array, handle multi-dimensional output for probability
    sv = shap_values[..., 1] if len(shap_values.shape) > 2 else shap_values

# Plot the summary
plt.figure(figsize=(10, 6))
shap.summary_plot(sv, X, plot_type="dot")
```
